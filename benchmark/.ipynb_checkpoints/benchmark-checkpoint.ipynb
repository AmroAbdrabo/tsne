## Benchmarks

Before running this script, ensure you have run the .py python scripts previously to generate the data

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline


input_sizes = np.arange(100, 4000, 150)
runs_per_input = 5
print(len(input_sizes))


def parse_file(filename: str):
    runtime = np.array([[0.0]*runs_per_input]*len(input_sizes)) # (5 runs for each input size), 30 is the number of different input sizes
    count = 0
    loc_cnt = 0
    f = open(filename, "r")
    for x in f:
        runtime[count//runs_per_input, loc_cnt%runs_per_input] = float(x)
        #print(runtime[count//runs_per_input, loc_cnt%runs_per_input])
        loc_cnt  = loc_cnt+1
        count = count+1
    return runtime
            
    
runtimeperp = parse_file("perp_runtimes.txt") # name of file used as argument to benchComputeGaussianPerp
runtimesed = parse_file("sedtimes.txt")
runtimezmean = parse_file("zmeantimes.txt")
runtimegrad = parse_file("gradtimes.txt")
```

```python
# gather median, mean, and std-dev for each input size

mean_per_input_perp = np.mean(runtimeperp, axis = 1) 
median_per_input_perp = np.median(runtimeperp, axis = 1)
stddev_per_input_perp = np.std(runtimeperp, axis = 1)
print(mean_per_input_perp.shape)

mean_per_input_sed = np.mean(runtimesed, axis = 1) 
median_per_input_sed = np.median(runtimesed, axis = 1)
stddev_per_input_sed = np.std(runtimesed, axis = 1)

mean_per_input_zmean = np.mean(runtimezmean, axis = 1) 
median_per_input_zmean = np.median(runtimezmean, axis = 1)
stddev_per_input_zmean = np.std(runtimezmean, axis = 1)

mean_per_input_grad = np.mean(runtimegrad, axis = 1) 
median_per_input_grad = np.median(runtimegrad, axis = 1)
stddev_per_input_grad = np.std(runtimegrad, axis = 1)
```

```python
# plot the data

def plotmarks(mean_per_input, stddev_per_input, title, median_per_input =None):
    fig, ax = plt.subplots(2, 2)

    fig.set_size_inches(15, 11)
    fig.suptitle(title, fontsize=16)

    # START OFF FIRST SUBPLOT TOP LEFT

    ax[0, 0].plot(input_sizes, mean_per_input, color = (0., 0., 0., 0.5), marker = 'o')
    ax[0, 0].set_facecolor((0.0, 0.2, 1., 0.07))
    ax[0, 0].spines['left'].set_visible(False)
    ax[0, 0].spines['top'].set_visible(False)
    ax[0, 0].spines['right'].set_visible(False)
    #ax.axis('off')

    a = ax[0,0].get_xgridlines()
    [b.set_color('white') for b in a]

    a = ax[0,0].get_ygridlines()
    [b.set_color('white') for b in a]


    ax[0, 0].grid()
    number =  8
    ax[0, 0].set_xlabel('N ~ input size', fontname = 'Helvetica', fontsize=17)

    ax[0, 0].set_title("       Mean runtime" + " on Intel i5-7360U\t"+ r"$\bf{" + "(cycles\ in\ TSC)" + "}$" , fontname='Helvetica', loc = 'left', fontsize=14)

    # ADD ERROR SHADOW TOP RIGHT

    ax[0, 1].fill_between(input_sizes, mean_per_input-stddev_per_input, mean_per_input+stddev_per_input, color = (0., 0., 0., 0.3))
    ax[0, 1].plot(input_sizes, mean_per_input, color = (0., 0., 0., 0.9), marker = 'o')
    ax[0, 1].set_facecolor((0.0, 0.2, 1., 0.07))
    ax[0, 1].spines['left'].set_visible(False)
    ax[0, 1].spines['top'].set_visible(False)
    ax[0, 1].spines['right'].set_visible(False)


    a = ax[0,1].get_xgridlines()
    [b.set_color('white') for b in a]

    a = ax[0,1].get_ygridlines()
    [b.set_color('white') for b in a]


    ax[0, 1].grid()
    number =  8
    ax[0, 1].set_xlabel('N ~ input size', fontname = 'Helvetica', fontsize=17)

    ax[0, 1].set_title("       Mean runtime" + " on Intel i5-7360U\t"+ r"$\bf{" + "(cycles\ in\ TSC)" + "}$" , fontname='Helvetica', loc = 'left', fontsize=14)


    # PLOT THE SAME PLOTS WITH MEDIAN RUNTIMES


    ax[1, 0].plot(input_sizes, median_per_input, color = (0., 0., 0., 0.5), marker = 'o')
    ax[1, 0].set_facecolor((1.0, 0.2, 0., 0.07))
    ax[1, 0].spines['left'].set_visible(False)
    ax[1, 0].spines['top'].set_visible(False)
    ax[1, 0].spines['right'].set_visible(False)
    #ax.axis('off')

    a = ax[1,0].get_xgridlines()
    [b.set_color('white') for b in a]

    a = ax[1,0].get_ygridlines()
    [b.set_color('white') for b in a]


    ax[1, 0].grid()
    number =  8
    ax[1, 0].set_xlabel('N ~ input size', fontname = 'Helvetica', fontsize=17)

    ax[1, 0].set_title("       Median runtime" + " on Intel i5-7360U\t"+ r"$\bf{" + "(cycles\ in\ TSC)" + "}$" , fontname='Helvetica', loc = 'left', fontsize=14)

    # ADD ERROR SHADOW TOP RIGHT

    ax[1, 1].fill_between(input_sizes, median_per_input-stddev_per_input, median_per_input+stddev_per_input, color = (0., 0., 0., 0.3))
    ax[1, 1].plot(input_sizes, median_per_input, color = (0., 0., 0., 0.9), marker = 'o')
    ax[1, 1].set_facecolor((1.0, 0.2, 0., 0.07))
    ax[1, 1].spines['left'].set_visible(False)
    ax[1, 1].spines['top'].set_visible(False)
    ax[1, 1].spines['right'].set_visible(False)


    a = ax[1,1].get_xgridlines()
    [b.set_color('white') for b in a]

    a = ax[1,1].get_ygridlines()
    [b.set_color('white') for b in a]


    ax[1, 1].grid()
    number =  8
    ax[1, 1].set_xlabel('N ~ input size', fontname = 'Helvetica', fontsize=17)

    ax[1, 1].set_title("      Median runtime" + " on Intel i5-7360U\t"+ r"$\bf{" + "(cycles\ in\ TSC)" + "}$" , fontname='Helvetica', loc = 'left', fontsize=14)

plotmarks(mean_per_input_perp, 1.96*stddev_per_input_perp, "Perplexity runtime",median_per_input_perp) #95% confidence interval

```

```python
plotmarks(mean_per_input_sed, 1.96*stddev_per_input_sed, "Square Euclidean Dist. runtime",median_per_input_sed)
```

```python
plotmarks(mean_per_input_zmean, 1.96*stddev_per_input_zmean, "Zero mean runtime",median_per_input_zmean)
```

```python
plotmarks(mean_per_input_grad, 1.96*stddev_per_input_grad, "Gradient update runtime",median_per_input_grad)
```

## Performance

### Flop counts estimated generously in the case of while loops or mutually exclusive conditionals

```python
# 400N^2 + 200N^2 + 600N^2 + 400N + 200N + 400N + 5N^2 
inputsquared = input_sizes**2
flop_count_gp = 1205*inputsquared + 1000*input_sizes

# (4N^2*D)/2 where D = input dimension = 16
flop_count_sed = 32*inputsquared

# (N^2 *3) + (N^2 * 3) + (N^2 *out_dim*3) + N*out_dim  + 5 *N*out_dim where out_dim = 2
flop_count_gradupd = 12*inputsquared + 12*input_sizes

# N*D + D + N*D where D = 16
flop_count_zmean = 16+ 32*input_sizes
```

```python
def plotperf(flops, mean_cycles_runtime, title):
    fig, ax = plt.subplots()

    fig.set_size_inches(11.2, 5.5)
    fig.suptitle(title, fontsize=16)

    # START OFF FIRST SUBPLOT TOP LEFT

    ax.plot(input_sizes, flops/mean_cycles_runtime, color = (0., 0., 0., 0.5), marker = 'o')
    ax.set_facecolor((0.0, 0.2, 1., 0.07))
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    #ax.axis('off')

    a = ax.get_xgridlines()
    [b.set_color('white') for b in a]

    a = ax.get_ygridlines()
    [b.set_color('white') for b in a]


    ax.grid()
    number =  8
    ax.set_xlabel('N ~ input size', fontname = 'Helvetica', fontsize=17)

    ax.set_title("       Mean performance" + " on Intel i5-7360U\t"+ r"$\bf{" + "(flops/cycles)" + "}$" , fontname='Helvetica', loc = 'left', fontsize=14)
```

```python
plotperf(flop_count_gp, mean_per_input_perp, "Performance of Gaussian Perplexity")
```

```python
plotperf(flop_count_sed, mean_per_input_sed, "Performance of Square Euclidean Dist.")
```

```python
plotperf(flop_count_gradupd, mean_per_input_grad, "Performance of Gradient Update")
```

```python
plotperf(flop_count_zmean, mean_per_input_zmean, "Performance of Zero Mean")
```
