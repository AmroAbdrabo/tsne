## Benchmark *computeGaussianPerplexity*

Before running this script, ensure you have run the benchComputeGP.cpp previously to generate the data

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline


input_sizes = np.arange(100, 4000, 150)
runs_per_input = 20
print(len(input_sizes))


def parse_file(filename: str):
    runtime = np.array([[0.0]*runs_per_input]*len(input_sizes)) # in reality this 5 will be 20 (20 runs for each input size), 30 is the number of different input sizes
    count = 0
    loc_cnt = 0
    f = open(filename, "r")
    for x in f:
        runtime[count//runs_per_input, loc_cnt%runs_per_input] = float(x)
        #print(runtime[count//runs_per_input, loc_cnt%runs_per_input])
        loc_cnt  = loc_cnt+1
        count = count+1
    return runtime
            
    
runtime = parse_file("perp_runtimes.txt") # name of file used as argument to benchComputeGaussianPerp
```

```python
# gather median, mean, and std-dev for each input size

mean_per_input = np.mean(runtime, axis = 1) 
median_per_input = np.median(runtime, axis = 1)
stddev_per_input= np.std(runtime, axis = 1)
```

```python
# plot the data

fig, ax = plt.subplots(2, 2)

fig.set_size_inches(15, 11)

# START OFF FIRST SUBPLOT TOP LEFT

ax[0, 0].plot(input_sizes, mean_per_input, color = (0., 0., 0., 0.5), marker = 'o')
ax[0, 0].set_facecolor((0.0, 0.2, 1., 0.07))
ax[0, 0].spines['left'].set_visible(False)
ax[0, 0].spines['top'].set_visible(False)
ax[0, 0].spines['right'].set_visible(False)
#ax.axis('off')

a = ax[0,0].get_xgridlines()
[b.set_color('white') for b in a]

a = ax[0,0].get_ygridlines()
[b.set_color('white') for b in a]


ax[0, 0].grid()
number =  8
ax[0, 0].set_xlabel('N ~ input size', fontname = 'Helvetica', fontsize=17)

ax[0, 0].set_title("Mean runtime" + " on Intel i5-7360U\t"+ r"$\bf{" + "(cycles\ in\ TSC)" + "}$" , fontname='Helvetica', loc = 'left', fontsize=14)

# ADD ERROR SHADOW TOP RIGHT

ax[0, 1].fill_between(input_sizes, mean_per_input-stddev_per_input, mean_per_input+stddev_per_input, color = (0., 0., 0., 0.3))
ax[0, 1].plot(input_sizes, mean_per_input, color = (0., 0., 0., 0.9), marker = 'o')
ax[0, 1].set_facecolor((0.0, 0.2, 1., 0.07))
ax[0, 1].spines['left'].set_visible(False)
ax[0, 1].spines['top'].set_visible(False)
ax[0, 1].spines['right'].set_visible(False)


a = ax[0,1].get_xgridlines()
[b.set_color('white') for b in a]

a = ax[0,1].get_ygridlines()
[b.set_color('white') for b in a]


ax[0, 1].grid()
number =  8
ax[0, 1].set_xlabel('N ~ input size', fontname = 'Helvetica', fontsize=17)

ax[0, 1].set_title("Mean runtime" + " on Intel i5-7360U\t"+ r"$\bf{" + "(cycles\ in\ TSC)" + "}$" , fontname='Helvetica', loc = 'left', fontsize=14)


# PLOT THE SAME PLOTS WITH MEDIAN RUNTIMES


ax[1, 0].plot(input_sizes, median_per_input, color = (0., 0., 0., 0.5), marker = 'o')
ax[1, 0].set_facecolor((1.0, 0.2, 0., 0.07))
ax[1, 0].spines['left'].set_visible(False)
ax[1, 0].spines['top'].set_visible(False)
ax[1, 0].spines['right'].set_visible(False)
#ax.axis('off')

a = ax[1,0].get_xgridlines()
[b.set_color('white') for b in a]

a = ax[1,0].get_ygridlines()
[b.set_color('white') for b in a]


ax[1, 0].grid()
number =  8
ax[1, 0].set_xlabel('N ~ input size', fontname = 'Helvetica', fontsize=17)

ax[1, 0].set_title("Median runtime" + " on Intel i5-7360U\t"+ r"$\bf{" + "(cycles\ in\ TSC)" + "}$" , fontname='Helvetica', loc = 'left', fontsize=14)

# ADD ERROR SHADOW TOP RIGHT

ax[1, 1].fill_between(input_sizes, median_per_input-stddev_per_input, median_per_input+stddev_per_input, color = (0., 0., 0., 0.3))
ax[1, 1].plot(input_sizes, median_per_input, color = (0., 0., 0., 0.9), marker = 'o')
ax[1, 1].set_facecolor((1.0, 0.2, 0., 0.07))
ax[1, 1].spines['left'].set_visible(False)
ax[1, 1].spines['top'].set_visible(False)
ax[1, 1].spines['right'].set_visible(False)


a = ax[1,1].get_xgridlines()
[b.set_color('white') for b in a]

a = ax[1,1].get_ygridlines()
[b.set_color('white') for b in a]


ax[1, 1].grid()
number =  8
ax[1, 1].set_xlabel('N ~ input size', fontname = 'Helvetica', fontsize=17)

ax[1, 1].set_title("Median runtime" + " on Intel i5-7360U\t"+ r"$\bf{" + "(cycles\ in\ TSC)" + "}$" , fontname='Helvetica', loc = 'left', fontsize=14)

```
