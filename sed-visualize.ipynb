```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# N vs. input size
in_N = 2 ** np.arange(5, 15, 1)
in_size = 8*(in_N*in_N+2*in_N)/(1024) # KB
fig, ax = plt.subplots(figsize=(12, 6))
sns.pointplot(in_N, in_size, ax = ax)
ax.hlines([32, 256, 6*1024], xmin = 0, xmax = in_N.size-1, colors="r") # cache sizes in KB
ax.set_yscale("log", basey=2)
```
